```python
a = df[["milestones","is_top500","age_first_milestone_year","has_roundB","funding_rounds","avg_participants","has_roundA"]]
```

```python
b=df[['status']]
def converter(status):
    if status == "closed":
        return 0
    
    else:
        return 1
b['status']=b['status'].apply(converter)
#concatenate the dataframes
df=pd.concat([a,b],axis=1)
```

```python
df.head()
```

```python
x = df.drop("status", axis=1)
y = df["status"]
```

```python
from imblearn.over_sampling import SMOTE
```

```python
oversample = SMOTE()
x,y=oversample.fit_resample(x,y)
```

Setting train and test dataset

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=49)
```

USING LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(class_weight='balanced')
```

```python
lr.fit(x_train,y_train)
```

```python
lr.predict(x_test)
pred=lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,pred))
```

USING SVM

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.svm import SVC
```

```python
from sklearn.metrics import accuracy_score, precision_score, f1_score, recall_score, confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
clf= BaggingClassifier(base_estimator = SVC(), n_estimators = 10, random_state = 0).fit(x_train, y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
tn, fp, fn, tp = confusion_matrix(list(y_test), list(y_pred), labels=[0,1]).ravel()

print('True Negative:', tn)
print('False Positive:', fp)
print('False Negative:', fn)
print('True Positive:', tp)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
```

```python
print(accuracy_score(y_test, y_pred))
```

```python
print(precision_score(y_test, y_pred))
```

```python
print(recall_score(y_test, y_pred))
```

```python
print(f1_score(y_test, y_pred))
```
